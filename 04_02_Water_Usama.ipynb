```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Proses pembacaan data dari Sumber
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data.head()
```

```python
#drop data yang tidak diperlukan perlu
data=data.drop('no', axis = 1)
data.head()
```

```python
#Lakukan Visualisasi
#Kita plot biar datanya terlihat

%matplotlib inline
ax1 = data.plot.scatter(x='x',y='y',c='DarkBlue')
```

### Lakukan segmentasi data dengan metode Clustering, metode yang digunakan bebas. Hasil akhir akan dievaluasi menggunakan fungsi silhouette score seperti contoh di bawah ini

```python
#Lakukan Proses Pencarian CLuster terbaik untuk data diatas

#START CODE
# Mengambil data x dan y dari DataFrame Anda
X = data[['x', 'y']].values

# Inisialisasi variabel untuk menyimpan nilai terbaik
best_score = -1
best_k = None

# Coba berbagai jumlah cluster dari 2 hingga sejumlah data - 1
for k in range(2, 30):
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, labels)

    # Membandingkan dengan nilai Silhouette Score terbaik yang telah ditemukan
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_k = k

# Menampilkan jumlah cluster terbaik dan nilai Silhouette Score terbaik
print("Jumlah Cluster Terbaik (Silhouette Score):", best_k)
print("Silhouette Score Terbaik:", best_score)
#END CODE
```

```python
#Lengkapi Code di bawah ini sehingga menghasilkan evaluasi jumlah cluster terbaik
#Contoh pemanggilan Library CLustering
#from sklearn.cluster import KMeans

#START CODE
# Inisialisasi variabel untuk menyimpan nilai terbaik
k = 15

# Coba jumlah cluster dari 2 hingga sejumlah data - 1
kmeans = KMeans(n_clusters=k, random_state=0)
label = kmeans.fit_predict(X)
#END CODE

score = silhouette_score(data, label)
print("silhouette score=",score)
```

## Proses Visualisasi hasil CLuster

```python
#Proses Visualisasi hasil cluster

#Pertama, masukkan dahulu label hasil clustering ke dalam data

result = data.copy()

#START CODE
result['cluster'] = label
result.head()
#END CODE
```

```python
#Lakukan Plotting terhadap data
import seaborn as sns

#Gunakan library seaborn, lebih mudah.

#START CODE
num_clusters = 15

# Menghasilkan palet warna menggunakan np.linspace untuk mendapatkan berbagai warna
colors = plt.cm.viridis(np.linspace(0, 1, num_clusters))

# Membuat palet warna custom
custom_palette = sns.color_palette(colors)

# Sekarang Anda dapat menggunakannya dalam sns.relplot
sns.set(style="whitegrid")  # Opsi ini mengatur gaya plot menjadi whitegrid
sns.relplot(data=result, x='x', y='y', hue='cluster', palette=custom_palette, markers=["o", "s", "D"])
plt.title('Visualisasi Hasil Clustering dengan 15 Cluster')

plt.show()
#END CODE
```
